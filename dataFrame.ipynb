concat dataframe

```python
indian_weather=pd.DataFrame({
    'city':['chennai','mumbai','bombay'],
    'temp':[32,33,34],
    'hum':[80,60,70]
})
indian_weather
```

```python
us_weather=pd.DataFrame({
    'city':['newyork','chicago','orlando'],
    'temp':[21,14,35],
    'hum':[68,65,75]
})
us_weather
```

```python
df=pd.concat([indian_weather,us_weather],keys=["india","us"])
df
```

```python
temp=pd.DataFrame({
    "city":["chennai","mumbai","bombay"],
    "temp":[32,34,33]
})
temp
```

```python
hum=pd.DataFrame({
    "city":["chennai","mumbai","bombay"],
    "hum":[80,70,60]
})
hum
```

```python
df=pd.concat([temp,hum],axis=1)
df
```

```python
df=pd.merge(temp,hum,on="city",suffixes=('_left','_right'))
df
```
